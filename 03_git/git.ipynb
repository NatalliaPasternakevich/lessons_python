# Git - система контроля версий
https://git-scm.com/ - скачать

*В MAC уже есть?*

### Инициализация репозитория - один раз (при создании репозитория)
если клонируем репозиторий - `git init` делать не надо.
```
cd /path/to/your/existing/code
git init
```
### Клонирование репозитория
Свой репозиторий (на новый компьютер) или сторонний проект себе. `git init` делать не надо.
При этом скачивается репозиторий, создается папка. Вытягивается вся структура репозитория.
```
git clone <repo url>
```
### Сохранение изменений
```bash
cd /path/to/project
echo "test content for git tutorial" >> CommitTest.txt # запись данных в файл.
git add CommitTest.txt # добавляем этот файл в будущий коммит
# или 
git add * # добавляем все файлы 
git add --all # добавляем все файлы
# или добавляем файл/директорию
git add <file>
git add <directory>
git commit -m "added CommitTest.txt to the repo" # делаем коммит (сохраняем изменения
```
## Удаленный репозиторий
```bash
git config --global user.name <name>             # Глобально установить имя пользователя (заменить <name>)
git config --local user.email <email>            # Глобольно установить email (заменить <email>)
git remote add <remote_name> <remote_repo_url>   # <remote_name> - обычно = origin
git push -u <remote_name> <local_branch_name>
# Отправить изменения на удаленный репозиторий где, 
# <remote_name> - обычно = origin, <local_branch_name> = master
git pull # скачать и объединить удаленный репозиторий с локальным.
```
### Конфигурация удаленных репозиториев
``` bash
git remote -v                           # посмотреть настройки
git remote add <name> <url>             # добавить удаленный репозиторий (при клоне добавляется сам) 
git remote rm <name> # re - remove      # соотв. удалить удаленный репозиторий
git remote rename <old-name> <new-name> # переименовать удаленный репозиторий
```
## Статус
```
git status
git log
```
## Совместная работа
```bash
git branch                     # list all branches
git branch <branch>            # create new one
git branch -d <branch>         # delete one
git checkout branch            # checkout to existing branch
git checkout -b <new-branch>   # create and checkout 
```
### Merging
#### Проверь HEAD! 
```bash
git merge <branch name>       # команда выполняется ИЗ ПРИНИМАЮЩЕЙ ветки, <branch name> - отдающая ветка
```
#### Конфликты
```
here is some content not affected by the conflict
<<<<<<< master
this is conflicted text from master
=======
this is conflicted text from feature branch
```
